```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def main():
    randoms = np.random.triangular(1, 5, 100, 1000)
    plt.hist(randoms, bins=100, color='blue')
    
    plt.xlabel("Number")
    plt.ylabel("Frequency")
    
    plt.savefig("hist.pdf", format="pdf")
    plt.show()
    
```

```python
if __name__ == "__main__":
    main()
```
