```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create an array that ranges from 0 to 1 in 100 steps
myArray = np.linspace(0,1,100)
```

```python
def expFunc(x):
    return np.exp(x)
```

```python
def main():
    y = expFunc(myArray)

    plt.figure()
    plt.plot(myArray,y)
    plt.xlabel("Time [milliseconds]")
    plt.ylabel("Awesomeness")
    plt.title("X VS. Y")
    plt.grid()
    plt.savefig("plot.pdf")
    plt.show()
```

```python
if __name__ == "__main__":
    main()
```
