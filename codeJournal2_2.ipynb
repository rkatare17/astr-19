```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,1,100)
```

```python
def sinx(x):
    return np.sin(x)

def cosx(x):
    return np.cos(x)
```

```python
def main():
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
    
    ax1.plot(x, sinx(x))
    ax1.set_title("Sin Graph")
    ax1.set_xlabel("x")
    ax1.set_ylabel("sin(x)")
    
    ax2.plot(x, cosx(x))
    ax2.set_title("Cos Graph")
    ax2.set_xlabel("x")
    ax2.set_ylabel("cos(x)")
    
    plt.savefig("plots.pdf", format="pdf")
    plt.show()
```

```python
if __name__ == "__main__":
    main()
```
