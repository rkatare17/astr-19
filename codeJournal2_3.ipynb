```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
randoms = np.random.uniform (0,1,1000)
```

```python
def main():
    plt.hist(randoms, bins = 100, color = 'red')
    plt.xlabel("Number")
    plt.ylabel("Frequency")
    plt.title("Histogram")
    
    plt.savefig("hist.pdf", format="pdf")
    plt.show()
```

```python
if __name__ == "__main__":
    main()
```
